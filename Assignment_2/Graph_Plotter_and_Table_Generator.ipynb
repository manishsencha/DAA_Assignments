```python
x = json.load(open('output.json'))
size_array = []
time_array = []
for i in x['data_1']:
    size_array.append(i['input_size'])
    time_array.append((i['time']))
```

```python
plt.plot(size_array, time_array)
plt.xlabel('Size of input')
plt.ylabel('Time taken (in microseconds)')
plt.savefig('plot_1.png')
plt.show()
```
