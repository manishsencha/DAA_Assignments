```python
import matplotlib.pyplot as plt
import json
```

```python
!g++ majority_element.cpp -o majority_element.exe
```

```python
!majority_element.exe
```

```python
data = json.load(open('output_majority_element.json'))
```

```python
size_array = []
time_array = []
```

```python
for i in data['data']:
    size_array.append(i['input_size'])
    time_array.append(i['time'])
```

```python
plt.plot(size_array, time_array)
plt.xlabel('Input Size')
plt.ylabel('Time (in microseconds)')
plt.title('Time vs Input Size')
plt.savefig('Majority_Element_Time_vs_Input_Size.png')
plt.show()
```
